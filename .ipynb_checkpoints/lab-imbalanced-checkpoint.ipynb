# LAB | Imbalanced

**Load the data**

In this challenge, we will be working with Credit Card Fraud dataset.

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv

Metadata

- **distance_from_home:** the distance from home where the transaction happened.
- **distance_from_last_transaction:** the distance from last transaction happened.
- **ratio_to_median_purchase_price:** Ratio of purchased price transaction to median purchase price.
- **repeat_retailer:** Is the transaction happened from same retailer.
- **used_chip:** Is the transaction through chip (credit card).
- **used_pin_number:** Is the transaction happened by using PIN number.
- **online_order:** Is the transaction an online order.
- **fraud:** Is the transaction fraudulent. **0=legit** -  **1=fraud**


```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
fraud = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv")
fraud.head()
```

**Steps:**

- **1.** What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset?
- **2.** Train a LogisticRegression.
- **3.** Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric.
- **4.** Run **Oversample** in order to balance our target variable and repeat the steps above, now with balanced data. Does it improve the performance of our model? 
- **5.** Now, run **Undersample** in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model?
- **6.** Finally, run **SMOTE** in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model? 

```python
import matplotlib.pyplot as plt
```

```python
# Check class distribution
print(fraud['fraud'].value_counts())
print(fraud['fraud'].value_counts(normalize=True))
```

```python
# Plot distribution
fraud['fraud'].value_counts().plot(kind='bar', title='Distribution of Fraudulent Transactions')
plt.show()
```

```python
#Train a Logistic Regression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
# Split features and target
X = fraud.drop('fraud', axis=1)
y = fraud['fraud']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train Logistic Regression (consider class_weight for imbalanced data)
model = LogisticRegression(class_weight='balanced', random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
#Evaluate the model
```

```python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
```

```python
y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:,1]
```

```python
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Precision/Recall: Important if detecting fraud is critical.

ROC-AUC: Measures ability to rank fraud vs legit.

```python
#Oversampling (RandomOverSampler)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)
```

```python
# Split again
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(X_res, y_res, test_size=0.3, random_state=42, stratify=y_res)
```

```python
# Scale features
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_res_scaled = scaler.transform(X_test_res)
```

```python
# Train logistic regression
model_res = LogisticRegression(random_state=42)
model_res.fit(X_train_res_scaled, y_train_res)
```

```python
# Evaluate
y_pred_res = model_res.predict(X_test_res_scaled)
y_proba_res = model_res.predict_proba(X_test_res_scaled)[:,1]
```

```python
print(classification_report(y_test_res, y_pred_res))
print("ROC-AUC:", roc_auc_score(y_test_res, y_proba_res))
```

```python
#Undersampling (RandomUnderSampler)
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)
```

```python
# Split, scale, train, evaluate (same steps as oversample)
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(X_res, y_res, test_size=0.3, random_state=42, stratify=y_res)
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_res_scaled = scaler.transform(X_test_res)
```

```python
model_res = LogisticRegression(random_state=42)
model_res.fit(X_train_res_scaled, y_train_res)
```

```python
y_pred_res = model_res.predict(X_test_res_scaled)
y_proba_res = model_res.predict_proba(X_test_res_scaled)[:,1]
```

```python
print(classification_report(y_test_res, y_pred_res))
print("ROC-AUC:", roc_auc_score(y_test_res, y_proba_res))
```

```python
#SMOTE (Synthetic Minority Oversampling)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
# Split, scale, train, evaluate (same as above)
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(X_res, y_res, test_size=0.3, random_state=42, stratify=y_res)
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_res_scaled = scaler.transform(X_test_res)
```

```python
model_res = LogisticRegression(random_state=42)
model_res.fit(X_train_res_scaled, y_train_res)
```

```python
y_pred_res = model_res.predict(X_test_res_scaled)
y_proba_res = model_res.predict_proba(X_test_res_scaled)[:,1]
```

```python
print(classification_report(y_test_res, y_pred_res))
print("ROC-AUC:", roc_auc_score(y_test_res, y_proba_res))
```
